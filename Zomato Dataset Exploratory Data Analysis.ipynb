## Zomato Dataset Exploratory Data Analysis 

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn  as sns
%matplotlib inline
```

```python
df = pd.read_csv('zomato.csv' ,encoding = 'latin-1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### in Data Analysis What All THing We Do 
1. Missisng Values
2. Explore about the numerical variable
3. explore about categorical variable
4. finding RElationship Between features 

```python
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
sns.heatmap(df.isnull(),yticklabels = False, cbar = False , cmap ='viridis')
plt.show()
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
## to check Data types
final_df.dtypes
```

```python
final_df.info()
```

```python
Country_names = final_df.Country.value_counts().index
Country_names
```

```python
Country_val = final_df.Country.value_counts().values
```

```python
## pie Chart Top 3 countries 
pie = plt.pie(Country_val[:3],labels= Country_names[:3],autopct='%1.2f%%')
plt.show(pie)
```

##### Observation : Zomato maximum records or transcation  are from India After that USA and Then UK

### Numerical Variables 

```python
final_df.columns
```

```python
 rating = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating count'})
```

```python
rating
```

### Observation 
1. when Rating bewteen   4.5 to 4.9      --> Excellemnt
2. when Rating between   4.0  to 4.4     --> Very Good
3.  when Rating between  3.5 to 3.9      -->  Good
4.  when Rating between  3.5 to 3.9      -->  Average
5.  when Rating between  1.8 to 2.4      -->  Poor

```python
rating.head()
```

```python
import matplotlib 
matplotlib.rcParams ['figure.figsize'] = (14, 7)
Rating_chart= sns.barplot(x="Aggregate rating", y ="Rating count",data=rating)
plt.show(Rating_chart)
```

```python
#change color
Rating_chart= sns.barplot(x="Aggregate rating", y ="Rating count",hue= "Rating color",data=rating,palette=['black','red','orange','yellow','green','green'])
```

```python
plt.show(Rating_chart)
```

### Observation :
1. Not Rated count is very high
2. Maxmimum number of rating are between 2.5 to 3.4

```python
## Count Plot 
count_plot = sns.countplot(x= 'Rating color', data = rating,hue= "Rating color",palette=['black','red','orange','yellow','green','green'])
```

```python
plt.show(count_plot)
```
